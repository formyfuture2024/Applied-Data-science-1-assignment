```python
# 3. Box Plot: Distribution of Time by Airline

# Extracting unique airline names from airline column

airlines = data['Airline'].unique()

# Creating a list of Time values corresponding to each airline

time_data = [data[data['Airline'] == airline]['Time'] for airline in airlines]

# Setting figure size and adding the labels

plt.figure(figsize=(10, 6))

# adding color to the boxplot 

box = plt.boxplot(time_data, labels=airlines, patch_artist=True, boxprops=dict(facecolor='skyblue', color='violet'),
                  whiskerprops=dict(color='black', linewidth=2), capprops=dict(color='orange', linewidth=2), 
                  flierprops=dict(markerfacecolor='red', marker='o', markersize=15),
                  medianprops=dict(color='brown', linewidth=2))

# adding the labels

plt.title('Distribution of Flight Time by Airline')
plt.xlabel('Airline')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)

# Showing the plot
plt.show()
```
