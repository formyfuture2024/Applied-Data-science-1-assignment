```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = ("C:/Users/subha/OneDrive/Desktop/airlines_delay.csv")
data = pd.read_csv(file_path)
```

```python
# Display column names and first few rows
print("Columns in the dataset:", data.columns)
print(data.head())
```

```python
#decribing the function using statistical method
data.describe()
```

```python
# 1. Pie Chart: Proportion of Flights by Airline
airline_counts = data['Airline'].value_counts()

#set figure size for better view
plt.figure(figsize=(8, 8))

#plotting a pie chart
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))

#Add title in piechart
plt.title('Proportion of Flights by Airline')

#display the pie chart
plt.show()
```

```python
# 2. Scatter Plot: Relationship Between Time and Length of Flight

#set figure size for better view
plt.figure(figsize=(8, 6))

#making a scatter plot using matplotlib
plt.scatter(data['Time'], data['Length'], color='purple')

#Add xlable,title,ylabel
plt.title('Flight Time vs Length')
plt.xlabel('Time (minutes)')
plt.ylabel('Length (distance)')

#showing the plot
plt.show()
```

```python
# 3. Box Plot: Distribution of Time by Airline

airlines = data['Airline'].unique()
time_data = [data[data['Airline'] == airline]['Time'] for airline in airlines]

plt.figure(figsize=(10, 6))
plt.boxplot(time_data, labels=airlines)
plt.title('Distribution of Flight Time by Airline')
plt.xlabel('Airline')
plt.ylabel('Time (minutes)')
plt.xticks(rotation=45)
plt.show()
```
